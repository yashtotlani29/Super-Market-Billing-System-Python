```python
## Executes for every New Customer
print("-"*50)
print("Welcome to Super Market Billing System")
print("-"*50)

#Name and Phone Number of the Customers
name = input("Enter your Name: ")
phone = input("Enter your Phone Number: ")

#Dictionary of products and item_price
product_items = {
    "Milk":30.0,
    "Flour":35.0,
    "Sugar":25.0,
    "Rice":50.0,
    "Biscuits":10.0,
    "Chips":10.0,
    "Chocolates":35.0,
    "Coffee Powder":45.0,
    "Tea Powder":55.0,
    "Apple":100.0,
    "Banana":40.0,
    "Potatoes":60.0,
    "Tomatoes":75.0,
    "Juices":80.0,
    "Maggi":60.0
}
print("-"*50)
print("product_items and price")
print("-"*50)

for i, j in product_items.items():
    print(i, ":" ,j)
print("-"*50)

#Loop use for[amount,cart,discount,total_amount]
while True:
    amount = 0
    cart = {}
    discount = 0
    total_amount = 0

    while True:
        p_item = input("Enter Product you want to buy: ").title()
        quantity = float(input("Enter Quantity you want to buy: "))
        values = []
        values.append(product_items[p_item])
        values.append(quantity)
        values.append(quantity*product_items[p_item])
        cart[p_item] = values
        amount += product_items[p_item] * quantity
        repeat = input("Do you want to add more items?")
        if repeat == "no":
            break

    #Calculate discount_amount
    if amount <= 500:
        discount = 0.05
        Percent = "discount 5%"
    elif amount <= 1500:
        discount = 0.10
        Percent = "discount 10%" 
    elif amount <= 3500:
        discount = 0.15
        Percent = "discount 15%"
    elif amount <= 5000:
        discount = 0.20
        Percent = "discount 20%"
    else:
        discount = 0
        Percent = "No discount"

    total_amount = amount - (amount * discount)
    print("-"*50)
    print("Name: ",name)
    print("Phone: ",phone)


    #Reciept Formatting
    print("{:<15}{:<11}{:<9}{:<12}".format("Item name","Price","Quantity","Total"))
    print ("-"*40)
    for i,j in cart.items():
        print("{:<15}{:<11}{:<9}{:<12}".format(i,j[0],j[1],j[0]*j[1]))
    print("-"*40)

    print("Final Amount: ",amount)
    print("Discount Applied: ",Percent)
    print("-"*50)

    print("Payable Amount: ",total_amount)
    print("-"*50)

    print("-"*50)
    print("\n\n")
    break
```
